Webscrapped as of 14/5/2024

```python
import pandas as pd 
import numpy as np 
import json

from app_store_scraper import AppStore
from google_play_scraper import app, Sort, reviews_all
```

```python
import requests
import bs4 
response = requests.get('https://apps.apple.com/cn/app/原神/id1467190251')
response.raise_for_status()
response.encoding = 'utf-8'

soup = bs4.BeautifulSoup(response.text, "html.parser")
```

Scrape reviews from China App store

```python
genshin_CN = AppStore(country='cn', app_name='原神'.encode('utf-8'), app_id='1467190251')
genshin_CN.review(how_many=2000)
```

```python
genshin_CN.reviews
```

```python
genshin_CN_df = pd.DataFrame(np.array(genshin_CN.reviews), columns=['review'])
genshin_CN_df2 = genshin_CN_df.join(pd.DataFrame(genshin_CN_df.pop('review').tolist()))
genshin_CN_df2.head()
```

Scrape reviews from US App store

```python
genshin_US = AppStore(country='us', app_name='genshin-impact', app_id='1517783697')
genshin_US.review(how_many=2000)
```

```python
genshin_US.reviews
```

```python
genshin_US_df = pd.DataFrame(np.array(genshin_US.reviews), columns=['review'])
genshin_US_df2 = genshin_US_df.join(pd.DataFrame(genshin_US_df.pop('review').tolist()))
genshin_US_df2.head()
```

Scrape reviews from playstore

```python
genshin_play = reviews_all(
    'com.miHoYo.GenshinImpact',
    sleep_milliseconds=0, # defaults to 0
    lang='en',
    country='sg',
    sort=Sort.NEWEST,
)
```

```python
genshin_play_df = pd.DataFrame(np.array(genshin_play), columns=['review'])
genshin_play_df2 = genshin_play_df.join(pd.DataFrame(genshin_play_df.pop('review').tolist()))
genshin_play_df2.head()
```

Convert to CSV

```python
genshin_CN_df2.to_csv('GenshinCN-AppStore-Reviews.csv', index=False)
genshin_US_df2.to_csv('GenshinUS-AppStore-Reviews.csv', index=False)
genshin_play_df2.to_csv('Genshin-PlayStore-Reviews.csv', index=False)
```
